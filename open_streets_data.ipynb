```python
import json
import geopandas as gpd

links = json.load(open('data/links.json'))
l = gpd.read_file('data/Open_Streets_Locations.csv')
nodes = l['segmentidt'].astype(str).str.lstrip('0').values
l = l[l['Borough Name'] == 'Manhattan']
nodes = [int(node) for node in nodes]

for feature in links['features']:
    if int(feature['properties']['SegmentID']) in nodes:
        print("Object ID:", feature['properties']['OBJECTID'])
        print("Segment ID:", feature['properties']['SegmentID'])
        print("Street:", feature['properties']['Street'])
        print("---")
```
